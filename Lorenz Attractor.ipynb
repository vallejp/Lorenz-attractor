# Lorenz Attractor

### Libraries:

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

System of equations: 

$\frac{dX}{dt}=f_1(X,Y,Z)=-\sigma X + \sigma Y$

$\frac{dY}{dt}=f_2(X,Y,Z)=rX - Y - XZ$

$\frac{dZ}{dt}=f_3(X,Y,Z)= XY - bZ$


Where $\sigma = \mu/k$ is the number of Prandtl, $r=R/R_c$ e $b=4/(1+a^2)$. 

```python
x, y, z = [1], [2], [3] #Initial conditions

, r, b = 10, 28, 8/3 #Parameters

time = 100000 

dt = 0.001 #footsteps

for i in range(time):
    x.append((-*x[-1] + *y[-1])*dt + x[-1])
    y.append((r*x[-1] - y[-1] -x[-1]*z[-1])*dt + y[-1])
    z.append((x[-1]*y[-1] - b*z[-1])*dt + z[-1])
    
mpl.rcParams['legend.fontsize'] = 10
fig = plt.figure(figsize=[15, 10])
ax = fig.gca(projection='3d')
ax.plot(x, y, z, linewidth=0.3)

plt.show()    
```
